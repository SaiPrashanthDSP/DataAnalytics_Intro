1. Generate a Brief Statistical Summary:
Load the SP500.csv file into a Pandas DataFrame.
Generate and display a brief statistical summary of the data, 
including key metrics such as mean, median, and standard deviation.

```python
import pandas as pd

# 1. Generate a brief statistical Summary
# Load File in to Pandas DataFrame.
df = pd.read_csv("SP500.csv")
# Getting mean median standard deviation of data
print(df.describe().loc[['mean','50%','std']])
```

2. Sort Daily Gain/Loss for January 2018:
Filter the data to include only transactions from January 2018.
Sort the daily gain/loss values in descending order.
Save the sorted data to a new CSV file

```python
# 2 . Sort Daily Gain or loss for 2018

# Reading the data from 2018 January month
jan_2018_month = df[df["Date"].astype(str).str[0:7] == "2018-01"]
gain_loss = jan_2018_month['Close'] - jan_2018_month['Open']

# Creating a data frame for gain or loss and sorting it
gain_loss_2018 = pd.DataFrame(gain_loss, columns=["Gain/Loss"])
gain_loss_2018.sort_values("Gain/Loss", axis=0, ascending=False, inplace=True)

# Display the sorted data
print(gain_loss_2018)

# Reading the sorted data to CSV
gain_loss_2018.to_csv("new.csv", index=False)
```


5. Segment and Analyze Volumes:
Segment the volumes into five bins using the following ranges:
1. O to 20% of the maximum volume
2. 20 to 40% of the maximum volume
3. 40 to 60% of the maximum volume
4. 60 to 80% of the maximum volume
5. 80 to 100% of the maximum volume
Add a new column named Volume bin cut to map each volume to its respective bin.
Segment the volumes into four quantiles and add a new column named Volume bin qeut 
to map each volume to its respective quantile.
Count the number of entries in each bin and each quantile.

Select and display rows where the volume falls into a specific bin.
Select and display rows where the volume falls into a specific quantile.
Sort the DataFrame based on the Volume bin cut column.
Sort the DataFrame based on the Volume bin qout column.

```python
# Convert 'Volume' to numeric
df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')

# Maximum volume value
max_volume = df['Volume'].max()

# 1. Create bins based on volume percentages
bins = [0, 0.2 * max_volume, 0.4 * max_volume, 0.6 * max_volume, 0.8 * max_volume, max_volume]
labels = ['0-20%', '20-40%', '40-60%', '60-80%', '80-100%']
df['Volume bin cut'] = pd.cut(df['Volume'], bins=bins, labels=labels)

# 2. Segment volumes into four quantiles
df['Volume bin quantile'] = pd.qcut(df['Volume'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# 3. Count the number of entries in each bin and quantile
bin_counts = df['Volume bin cut'].value_counts()
quantile_counts = df['Volume bin quantile'].value_counts()

# Display bin counts
print("Bin Counts:")
print(bin_counts)

# Display quantile counts
print("\nQuantile Counts:")
print(quantile_counts)

# 4. Select rows in specific bins
bin_selected = df[df['Volume bin cut'] == '0-20%']
print(f"\nRows with Volume in '0-20%' bin:")
print(bin_selected)

# 5. Select rows in specific quantiles
quantile_selected = df[df['Volume bin quantile'] == 'Q1']
print(f"\nRows with Volume in 'Q1' quantile:")
print(quantile_selected)

# 6. Sort by volume bin cut
sorted_by_bin = df.sort_values(by='Volume bin cut')
print("\nData sorted by Volume bin cut:")
print(sorted_by_bin)

# 7. Sort by volume bin quantile
sorted_by_quantile = df.sort_values(by='Volume bin quantile')
print("\nData sorted by Volume bin quantile:")
print(sorted_by_quantile)
```

1. Monthly Report for 2017-2018:
Generate a report showing the monthly average of the open price, close price, transaction volume, and gain/loss.
Perform a query to find all months where the open price falls within a specified range.


```python
# Monthly Report for 2017 - 2018

# Filter valid dates and convert types
df = df[df['Date'].str.match(r'\d{4}-\d{2}-\d{2}')]
df[['Open', 'Close', 'Volume']] = df[['Open', 'Close', 'Volume']].apply(pd.to_numeric, errors='coerce')

# Convert 'Date' to datetime and extract YearMonth
df['Date'] = pd.to_datetime(df['Date'])
df['YearMonth'] = df['Date'].dt.to_period('M')

# Filter data for 2017-2018
df_filtered = df[(df['Date'].between('2017-01-01', '2018-12-31'))]

# Generate monthly report
monthly_report = df_filtered.groupby('YearMonth').agg(
    MonthlyAvgOpen=('Open', 'mean'),
    MonthlyAvgClose=('Close', 'mean'),
    MonthlyTotalVolume=('Volume', 'sum'),
    GainLoss=('Close', 'last')
)

# Subtract first open price to calculate GainLoss
monthly_report['GainLoss'] -= df_filtered.groupby('YearMonth')['Open'].first()

# Filter months where open price is between 2500 and 2600
months_in_range = df_filtered.loc[df_filtered['Open'].between(2500, 2600), 'YearMonth'].unique()

# Print results
print("Monthly Report for 2017-2018:")
print(monthly_report)

print("\nMonths with Open Price between 2500 and 2600:")
print(months_in_range)
```

2. Yearly Report from 1950 to 2018:
Create a report with the annual average of the open price, close price, transaction volume, and gain/loss for 
each year from 1950 to 2018.
Identify the most profitable year.

```python
#ANNUAL AVERAGE YEARLY REPORT FOR 1950 TO 2018.

# Convert 'Date' to string before applying .str functions (Fixes the error)
df = df[df['Date'].astype(str).str.match(r'\d{4}-\d{2}-\d{2}')]

# Convert necessary columns to numeric
df[['Open', 'Close', 'Volume']] = df[['Open', 'Close', 'Volume']].apply(pd.to_numeric, errors='coerce')

# Convert 'Date' column to datetime and extract Year
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Filter data for years 1950-2018
df_filtered = df[df['Year'].between(1950, 2018)]

# Generate yearly report
yearly_report = df_filtered.groupby('Year').agg(
    AnnualAvgOpen=('Open', 'mean'),
    AnnualAvgClose=('Close', 'mean'),
    AnnualTotalVolume=('Volume', 'sum')
)

# Compute GainLoss directly
yearly_report['GainLoss'] = df_filtered.groupby('Year')['Close'].last() - df_filtered.groupby('Year')['Open'].first()

# Find the most profitable year
most_profitable_year = yearly_report.loc[yearly_report['GainLoss'].idxmax()]

# Print results
print("Yearly Report from 1950 to 2018:")
print(yearly_report)

print("\nMost Profitable Year:")
print(most_profitable_year)

```

3. Every Five-Year Report from 1950 to 2018
Generate a report showing the average open price, close price, transaction volume, 
and gain/loss for every five-year period from 1950 to 2018.
Determine the most profitable five-year period.

```python
#EVERY FIVE YEAR AVERAGE REPORT
# Filter valid dates
df = df[df['Date'].astype(str).str.match(r'\d{4}-\d{2}-\d{2}')]

# Convert necessary columns to numeric
df[['Open', 'Close', 'Volume']] = df[['Open', 'Close', 'Volume']].apply(pd.to_numeric, errors='coerce')

# Convert 'Date' to datetime and extract Year
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# Filter data for the years 1950-2018
df_filtered = df[df['Year'].between(1950, 2018)]

# Create five-year groupings
df_filtered['FiveYearGroup'] = ((df_filtered['Year'] - 1950) // 5) * 5 + 1950

# Generate five-year report
five_year_report = df_filtered.groupby('FiveYearGroup').agg(
    FiveYearAvgOpen=('Open', 'mean'),
    FiveYearAvgClose=('Close', 'mean'),
    FiveYearTotalVolume=('Volume', 'sum')
)

# Compute GainLoss directly
five_year_report['GainLoss'] = df_filtered.groupby('FiveYearGroup')['Close'].last() - df_filtered.groupby('FiveYearGroup')['Open'].first()

# Find the most profitable five-year period
most_profitable_five_year = five_year_report.loc[five_year_report['GainLoss'].idxmax()]

# Print results
print("Every Five-Year Report from 1950 to 2018:")
print(five_year_report)

print("\nMost Profitable Five-Year Period:")
print(most_profitable_five_year)
```

```python
dataframe=pd.read_csv("Sacramentorealestatetransactions.csv")
# Grouping the data first by city name and then by type
grouped_data = dataframe.groupby(['city', 'type']).agg(
    AveragePrice=('price', 'mean'),
    TotalTransactions=('price', 'count')
).reset_index()
print(grouped_data)
```

```python
# Grouping the data by city and type and calculating the highest, median, and lowest transactions
grouped_data = dataframe.groupby(['city', 'type'])['price'].agg(
    HighestTransaction='max',
    MedianTransaction='median',
    LowestTransaction='min'
).reset_index()
print(grouped_data)
```

```python
# Grouping the data by zipcode and type and calculating the average transactions
grouped_data = dataframe.groupby(['zip', 'type'])['price'].mean().reset_index()
print(grouped_data)
```
