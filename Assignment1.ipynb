1. Write a Python program to count the number of strings in a list where each string has a length of 2 or more and is a palindrome. 
Sample: ['abdefc', 'xygyz', 'abeba', '123321'] 
output: 2

```python
def count_noof_palindromes(lists_of_palindrome):
    count=0
    original_list=lists_of_palindrome.copy()  ##to keep the original list 
    while original_list:
        new_string =original_list.pop(0) ## to remove and return the first element from the string
        if len(new_string)>=2 and new_string == new_string[::-1]: #to check if the length is 2 or more and to check if the string is equal to its reverse
            count+= 1
    return count
sample=['abdefc', 'xygyz', 'abeba', '123321'] # sample input 
result1 = count_noof_palindromes(sample)
print("Total_palindromes in the given list:",result1)
```

3. Write a Python program to split a list into a list of dictionaries, where each dictionary has an odd item as the key and the following even item as the value. 
Sample: ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n'] 
Output: [{‘a’:’b’},{‘c’:’d’},{‘e’:’f’},{‘g’:’h’},{‘i’:’j’},{‘k’:’l’},{‘m ’:‘n’}]

```python
def split_dictionary(listofdict):
    result = []   #empty list to store the disctionaries
    for i in range(0, len(listofdict) - 1, 2):  # looping through the list to take 2 values at once
        result.append({listofdict[i]: listofdict[i + 1]}) #dictionary with the first item as key and second item as value
    return result
sample = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n'] # sample input given as per the question
result = split_dictionary(sample)
print("list of dictionary:\n",result)
```

5. write a Python program to print all the common values in a dictionary.

```python

def check_common_values(common_values):
    no_of_values={}      # initialized an empty dictionary
    for value in common_values.values():
        no_of_values[value]= no_of_values.get(value,0)+1   #count values

    common_values = [value1 for value1, count in no_of_values.items() if count>1] #create a list of values that apper more than once

    if common_values:   
        print("common values:", common_values) #condition to check the common values
    else:
        print("no common values found.")

sample={"apples":10, "oranges":10, "grapes":20, "lemons":25, "strawberries":20, "pineapples":5} # sample input 
check_common_values(sample)
```

7. Write a Python program to remove tuples containing a specific value from a list of tuples.

```python

def remove_tuples(specificValue, value):
    checked_list=[]   #initialised an empy list to store the values after removing the tuples
    for tuples in specificValue:    
        if value not in tuples:      #to check the specific value in the tuples 
            checked_list.append(tuples)   #if condition meet than it add to the new checked list
    return checked_list

list_of_tuples=[(5,1,5),(1,6,9),(2,4,5),(3,6,6),(4,1,3),(2,3,4),(6,7,8)] #sample input
specific_value_removed=5  # specific value removed form the tuple

checked_list=remove_tuples(list_of_tuples, specific_value_removed) #list after removing the specific value from the list

print("final_list_after_removing_specific_values:",checked_list) #final list to print after removing the specific value
```

9. Use Yahoo Finance to download one month of daily stock data for three tickers: GOOG, AAPL, and TSLA, and store their prices in three separate lists.
Use list comprehension to identify the top three prices for each ticker and display these prices.

```python
#1.for this question We have first downloaded the data using the library 
#2.then we have choosen the to make a list of the closing prices for the each tickers 
#2.After that we havve given the top three prices for each ticker and displayed the prices 

import yfinance as yf
import pandas as pd

start_date= pd.Timestamp(year=2025,month=1, day=1)   #start date
end_date= pd.Timestamp(year=2025,month=2, day=1)      #end date 

tickers=['GOOG','AAPL','TSLA'] #list of the tickers as per the question
data=yf.download(tickers,start=start_date,end=end_date) #yf will download the respective data of each tickers as per start and end date

display(data)
```

```python
#to extract and print the close prices and convert it into list for each tickers (goog, aapl, tsla)
goog_prices =data['Close']['GOOG'].tolist()
aapl_prices=data['Close']['AAPL'].tolist()
tsla_prices=data['Close']['TSLA'].tolist()
print ("GOOG_prices:\n ",goog_prices)
print ("AAPL_prices:\n",aapl_prices)
print("TSLA_prices:\n",tsla_prices)
```

```python
def get_top_three(prices):
    return sorted(prices, reverse=True)[:3] # sorted the prices and sliced the sorted list to get the top three prices

top_goog_prices=get_top_three(goog_prices)
top_aapl_prices=get_top_three(aapl_prices)
top_tsla_prices=get_top_three(tsla_prices)

#printed the top three prices for the each tickers
print("GOOG TOP THREE PRICES:\n",top_goog_prices)
print("AAPL TOP THREE PRICES:\n",top_aapl_prices)
print("TSLA TOP THREE PRICES:\n",top_tsla_prices)
```
